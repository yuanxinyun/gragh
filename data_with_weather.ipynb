```python
import warnings, datetime
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt

import datetime
df = pd.read_pickle("2016_temp1.pkl")
coordinate = pd.read_pickle("coordinate_interpolate.pkl")
```

```python
df
```

```python
df = df.assign(O_LATITUDE=0.0,
               O_LONGITUDE=0.0,
               D_LATITUDE=0.0,
               D_LONGITUDE=0.0,
               O_WEATHER=pd.Series(dtype='object'),
               D_WEATHER=pd.Series(dtype='object'),
               O_TEMP=0.0,
               O_PRCP=0.0,
               O_WSPD=0.0,
               D_TEMP=0.0,
               D_PRCP=0.0,
               D_WSPD=0.0)
#预先分配空间速度会快一些
```

```python
df
```

```python
import time
start = time.time()


# 定义一个函数，根据输入的行数据获取经纬度和天气数据
def get_coordinate_and_weather(row):
    # 从coordinate数据框中获取对应的经纬度和天气数据
    row['O_LATITUDE'] = coordinate.loc[row['ORIGIN_INDEX'], 'LATITUDE']
    row['O_LONGITUDE'] = coordinate.loc[row['ORIGIN_INDEX'], 'LONGITUDE']
    row['D_LATITUDE'] = coordinate.loc[row['DEST_INDEX'], 'LATITUDE']
    row['D_LONGITUDE'] = coordinate.loc[row['DEST_INDEX'], 'LONGITUDE']
    row['O_WEATHER'] = coordinate.loc[row['ORIGIN_INDEX'], 'WEATHER_INTERPOLATE']
    row['D_WEATHER'] = coordinate.loc[row['DEST_INDEX'], 'WEATHER_INTERPOLATE']
    # 返回修改后的行数据
    return row

# 使用apply循环对每一行数据调用get_coordinate_and_weather函数，并将结果赋值给df
df = df.apply(get_coordinate_and_weather, axis=1)

#print("222")


import datetime

# 四舍五入到最近的整点
def rounder(t):
    if t.minute >= 30:
        # 增加一个小时
        return (t + datetime.timedelta(hours=1)).replace(second=0, microsecond=0, minute=0)
    else:
        return t.replace(second=0, microsecond=0, minute=0)


# 定义一个函数，根据输入的时间以及机场天气数据查询天气数据，并返回temp,prcp,wspd
def query_weather(time,weather):
    # 将时间四舍五入到最近的整点
    time = rounder(time)
    try:
        row = weather.loc[time]
        # 返回temp,prcp,wspd
        return row['temp'], row['prcp'], row['wspd']
    except KeyError:
        # 如果找不到对应的行，说明没有天气记录
        # print('没有天气记录')
        return None, None, None





# 定义一个函数，根据输入的行数据查询出发地和目的地的温度、降水量和风速
def get_temp_prcp_wspd(row):
    # 调用query_weather函数，传入出发时间和出发地天气数据
    row['O_TEMP'], row['O_PRCP'], row['O_WSPD'] = query_weather(row['CRS_DEP_TIME'], row['O_WEATHER'])
    # 调用query_weather函数，传入到达时间和目的地天气数据
    row['D_TEMP'], row['D_PRCP'], row['D_WSPD'] = query_weather(row['CRS_ARR_TIME'], row['D_WEATHER'])
    # 返回修改后的行数据
    return row

# 使用apply循环对每一行数据调用get_temp_prcp_wspd函数，并将结果赋值给df
df = df.apply(get_temp_prcp_wspd, axis=1)
end = time.time()
length = end - start
print("耗时", length, "秒")



```
